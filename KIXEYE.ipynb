```python
import argparse
import sys
from sklearn.cluster import KMeans
from math import sqrt

COUNTRIES = {
    "Balinese": "Bali",
    "Bolivian": "Bolivia",
    "Brazilian": "Brazil",
    "Costa Rican": "Costa Rica",
    "Dominican": "Dominican Republic",
    "Salvadorean": "El Salvador",
    "Ethiopian": "Ethiopia",
    "Guatemalan": "Guatemala",
    "Indian": "India",
    "Kenyan": "Kenya",
    "Malian": "Mali",
    "Mexican": "Mexico",
    "Panamanian": "Panama",
    "Peruvian": "Peru",
    "Sumatran": "Sumatra",
}

def parse(name):
    # to be implemented
    print name #print full name to double check
    #initialize 
    Decaf = False
    Organic = False
    Fair_trade = False
    Adjective = ''
    Country = ''
    words = name.split(' ')
    total_words = len(words)
    i = 0
    while i < total_words: #iterate through each word
        if words[i] == 'Decaf': 
            Decaf = True
        elif words[i] == 'Organic': 
            Organic = True
        elif words[i] == 'Fair' and words[i+1] == 'Trade': 
            Fair_trade = True
            i+=1
        elif words[i] in COUNTRIES.keys():
            Country = COUNTRIES[words[i]]
        elif words[i]+' '+words[i+1] in COUNTRIES.keys():
            Country = COUNTRIES[words[i]+' '+words[i+1]]
            i+=1
        else: Adjective += words[i]+' '
        i+=1
    Adjective = Adjective[:-1] #remove last space at end of adjective string
    print "\tDecaf\t\t%s" %Decaf
    print "\tOrganic\t\t%s" %Organic
    print "\tFair trade\t%s" %Fair_trade
    print "\tAdjective\t%s" %Adjective
    print "\tCountry\t\t%s" %Country

def summarize_parse(name):
    #initialize 
    Decaf = False
    Organic = False
    Fair_trade = False
    Adjective = ''
    Country = ''
    words = name.split(' ')
    total_words = len(words)
    i = 0
    while i < total_words: #iterate through each word
        if words[i] == 'Decaf': 
            Decaf = True
        elif words[i] == 'Organic': 
            Organic = True
        elif words[i] == 'Fair' and words[i+1] == 'Trade': 
            Fair_trade = True
            i+=1
        elif words[i] in COUNTRIES.keys():
            Country = COUNTRIES[words[i]]
        elif words[i]+' '+words[i+1] in COUNTRIES.keys():
            Country = COUNTRIES[words[i]+' '+words[i+1]]
            i+=1
        else: Adjective += words[i]+' '
        i+=1
    Adjective = Adjective[:-1] #remove last space at end of adjective string
    return Decaf, Organic, Fair_trade, Adjective, Country


def summarize(file):
    # to be implemented
    reviewer_ids = []
    total_coffee_types = []
    decaf_count = {}
    organic_count = {}
    fair_trade_count = {}
    adjectives_count = {}
    country_count = {}
    ratings = open(file, 'r')
    for line in ratings:
        fields = line.split('\t')
        reviewer = fields[0]
        if reviewer not in reviewer_ids: reviewer_ids.append(reviewer)
        coffee_name = fields[1]
        if coffee_name not in total_coffee_types: total_coffee_types.append(coffee_name)

    for coffee_name in total_coffee_types:
        decaf, organic, fair_trade, adjectives, country = summarize_parse(coffee_name)
        decaf_count[str(decaf)] = decaf_count.get(str(decaf),0) + 1
        organic_count[str(organic)] = organic_count.get(str(organic),0) + 1
        fair_trade_count[str(fair_trade)] = fair_trade_count.get(str(fair_trade),0) + 1
        words = adjectives.split(' ')
        for word in words:
            adjectives_count[str(word)] = adjectives_count.get(str(word),0) + 1
        country_count[str(country)] = country_count.get(str(country),0) + 1

    print "Total People %s" %len(reviewer_ids)
    print "Total coffee types %s" %len(total_coffee_types)
    print "Decaf\n  True %s" %decaf_count['True']
    print "  False %s" %decaf_count['False']
    print "Organic\n  True %s" %organic_count['True']
    print "  False %s" %organic_count['False']
    print "Fair trade\n  True %s" %fair_trade_count['True']
    print "  False %s" %fair_trade_count['False']
    print "Adjective"
    for word in adjectives_count.keys():
        print " ", word, adjectives_count[word] 
    print "Country" 
    for country in country_count.keys():
        print " ", country, country_count[country] 

def create_dataset(file):
    dataset = {}
    ratings = open(file, 'r')
    for line in ratings:
        line = line.rstrip()
        fields = line.split('\t')
        reviewer = fields[0]
        coffee = fields[1]
        score = fields[2]
        if reviewer not in dataset.keys(): 
            dataset[reviewer] = {}
            dataset[reviewer][coffee] = int(score)
        else:
            dataset[reviewer][coffee] = int(score)
    return dataset

def recommend(file):    
    dataset = create_dataset(file)
    def pearson_correlation(person1,person2):
        # To get both rated items
        both_rated = {}
        for item in dataset[person1]:
            if item in dataset[person2]:
                both_rated[item] = 1

        number_of_ratings = len(both_rated) 

        # Checking for number of ratings in common
        if number_of_ratings == 0:
             return 0

        # Add up all the preferences of each user
        person1_preferences_sum = sum([dataset[person1][item] for item in both_rated])
        person2_preferences_sum = sum([dataset[person2][item] for item in both_rated])

        # Sum up the squares of preferences of each user
        person1_square_preferences_sum = sum([pow(dataset[person1][item],2) for item in both_rated])
        person2_square_preferences_sum = sum([pow(dataset[person2][item],2) for item in both_rated])

        # Sum up the product value of both preferences for each item
        product_sum_of_both_users = sum([dataset[person1][item] * dataset[person2][item] for item in both_rated])

        # Calculate the pearson score
        numerator_value = product_sum_of_both_users - (person1_preferences_sum*person2_preferences_sum/number_of_ratings)
        denominator_value = sqrt((person1_square_preferences_sum - pow(person1_preferences_sum,2)/number_of_ratings) * (person2_square_preferences_sum -pow(person2_preferences_sum,2)/number_of_ratings))
        if denominator_value == 0:
            return 0
        else:
            r = numerator_value/denominator_value
            return r 
    
    # Gets recommendations for a person by using a weighted average of every other user's rankings
    totals = {}
    simSums = {}
    rankings_list =[]  
    total_reviewers = 0
    for reviewer in dataset.keys():
        for other in dataset:
        # don't compare me to myself
            if other == reviewer:
                continue
            sim = pearson_correlation(reviewer,other)

            # ignore scores of zero or lower
            if sim <=0:
                continue
            for item in dataset[other]:

            # only score movies i haven't seen yet
                if item not in dataset[reviewer] or dataset[reviewer][item] == 0:
                    # Similrity * score
                    totals.setdefault(item,0)
                    totals[item] += dataset[other][item]*sim
                    # sum of similarities
                    simSums.setdefault(item,0)
                    simSums[item]+= sim

        # Create the normalized list 
        rankings = [(total/simSums[item],item) for item,total in totals.items()]     
        rankings.sort()
#        rankings.reverse()
        print rankings[:3]
        total_reviewers+=1
    print total_reviewers

#     kmeans = KMeans(n_clusters=12)
#     kmeans.fit(data)
```
